```python
# Ejemplo 1: Aritmética para Presupuesto de Viaje
# En este ejemplo usamos operadores aritméticos para calcular gastos basados en lo que el usuario ingresa.
# Pedimos el número de días para definir las vueltas del ciclo
dias = int(input("¿Cuántos días durará tu viaje?: "))
total_gastos = 0

for i in range(1, dias + 1):
    # Solicitamos el gasto diario
    gasto = float(input(f"Gasto del día {i}: "))

    # OPERADOR ARITMÉTICO (+): Sumamos el gasto al acumulador
    total_gastos = total_gastos + gasto

    # EXPLICACIÓN: Cada dato ingresado se suma al valor anterior de 'total_gastos'
    # para obtener una suma acumulativa al final del ciclo.

print(f"El costo total del viaje es: ${total_gastos}")
```

```python
# Ejemplo 2: Relacionales para Control de Inventario
# Aquí usamos operadores de comparación para determinar si un producto necesita reabastecimiento.
productos = int(input("¿Cuántos artículos vas a registrar?: "))

for i in range(productos):
    nombre = input("Nombre del artículo: ")
    while True:
        try:
            stock = int(input(f"Cantidad actual de {nombre}: "))
            break # Exit the loop if input is a valid integer
        except ValueError:
            print("¡Entrada inválida! Por favor, ingresa un número entero para la cantidad.")

    # OPERADOR RELACIONAL (<): Comparamos el stock con el mínimo deseado
    if stock < 10:
        # EXPLICACIÓN: Si la comparación resulta verdadera (True), se ejecuta este bloque
        print(f"ALERTA: El artículo '{nombre}' tiene poco stock. ¡Comprar más!")
    else:
        print(f"Stock de '{nombre}' es suficiente.")
```

```python
# Ejemplo 3: Lógica para Becas Estudiantiles
# Este ejemplo combina operadores relacionales y lógicos (and) para validar si un estudiante califica para un beneficio.
aspirantes = int(input("¿Cuántos estudiantes aplican para la beca?: "))

for i in range(aspirantes):
    promedio = float(input("Ingresa el promedio (0-10): "))
    ingresos = float(input("Ingresa el ingreso mensual familiar: "))

    # OPERADORES RELACIONALES (>=, <) y LÓGICO (and)
    if promedio >= 9.0 and ingresos < 1500:
        # EXPLICACIÓN: El operador 'and' exige que AMBAS condiciones sean verdaderas.
        # El estudiante debe ser excelente académicamente Y tener necesidad económica.
        print("RESULTADO: Candidato seleccionado para la beca.")
    else:
        print("RESULTADO: No cumple con los requisitos combinados.")
```

```python
# Ejemplo 4: Módulo para Clasificación de Turnos
# Utilizamos el operador aritmético de residuo (%) para asignar turnos (pares/impares) a las personas en una fila.
personas = int(input("¿Cuántas personas hay en la fila?: "))

for turno in range(1, personas + 1):
    nombre = input(f"Nombre de la persona con turno {turno}: ")

    # OPERADOR ARITMÉTICO (%): Residuo de la división
    # OPERADOR RELACIONAL (==): Comparación de igualdad
    if turno % 2 == 0:
        # EXPLICACIÓN: Si el residuo de dividir el turno entre 2 es cero, es par.
        print(f"{nombre} va a la VENTANILLA A (Pares).")
    else:
        # EXPLICACIÓN: Si el residuo es distinto de cero, es impar.
        print(f"{nombre} va a la VENTANILLA B (Impares).")
```

```python
# Ejemplo 5: Lógica "OR" para Validación de Acceso
# En este último ejemplo, usamos el operador lógico or para permitir el acceso si se cumple al menos una de dos condiciones.
invitados = int(input("¿Cuántas personas intentan entrar?: "))

for i in range(invitados):
    tiene_invitacion = input("¿Tiene invitación física? (si/no): ").lower()
    esta_en_lista = input("¿Está en la lista VIP? (si/no): ").lower()

    # OPERADOR LÓGICO (or): Basta con que una condición sea cierta
    # OPERADOR RELACIONAL (==): Compara el texto ingresado
    if tiene_invitacion == "si" or esta_en_lista == "si":
        # EXPLICACIÓN: El operador 'or' devuelve True si cualquiera de los dos lados es 'si'.
        # Esto permite que alguien entre aunque haya olvidado su invitación física pero esté en lista.
        print("ACCESO PERMITIDO. Bienvenido al evento.")
    else:
        print("ACCESO DENEGADO. No cumple ningún requisito.")
```
